```python
# 1er Dataset qui vient de Kaggle
import zipfile
import os

# Dossier courant (où le notebook est lancé)
notebook_dir = os.getcwd()  

# Chemin vers le zip (dans data/ depuis la racine du projet)
zip_path = os.path.join(notebook_dir, "..", "data", "archive.zip")

# Dossier de destination : data/raw/
extract_to = os.path.join(notebook_dir, "..", "data", "raw")
os.makedirs(extract_to, exist_ok=True)

# Dézipper
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

print(f"✅ Dataset dézippé dans : {extract_to}")
```
